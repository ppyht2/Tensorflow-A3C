```python
from utils import *
import matplotlib.pyplot as plt
from scipy.misc import imresize
import gym

%matplotlib inline
```

# Preprocessing - Basic Resize

```python
env_name = 'MsPacman-v0'
env = gym.make(env_name)

obs = env.reset()

plt.figure(figsize=(18,9))
plt.subplot(1,2,1)
plt.imshow(obs)
plt.title('Origional')
plt.subplot(1,2,2)
new_image = imresize(obs, [84, 84])
plt.imshow(new_image)
plt.title('Down Sampled')
env.close()

print('New Image Shape:', new_image.shape)
print('New Image Type:', type(new_image))
```

# Init_state(s) - State intitalisation 

```python
env = gym.make(env_name)

obs = env.reset()
plt.imshow(obs)
plt.title('Observation')
plt.show()
plt.imshow(preprocess(obs))
plt.title('process')
plt.show()
state = init_state(obs)
disp_state(state)


print('State Shape:', state.shape)
print('State Type:', type(state))
env.close()
```

# update_state - State modification 

```python
env = gym.make(env_name)

obs = env.reset()
state = init_state(obs)
for i in range(120):
    obs, _, _, _ = env.step(env.action_space.sample())
    state = update_state(state, obs)
    

disp_state(state)


print('State Shape:', state.shape)
print('State Type:', type(state))
env.close()
```
