```python
r_hist = utils.load_pickle('return_hist')
print('History Length:', len(r_hist))
```

```python
t = []
g = []
gp = []
gs = []
for i in r_hist:
    t.append(i[0])
    g.append(i[1])
    gp.append(i[2][0])
    gs.append(i[2][1])
print(max(gs))
```

```python
diff = np.abs(np.asarray(t)-57959)
index = np.argmin(diff)
print(index)
print(t[1699])
#print(g[1699])
print(np.mean(g[1699]))
print(np.min(g[1699]))
print(np.max(g[1699]))


print(np.mean(gp[1699]))
print(np.min(gp[1699]))
print(np.max(gp[1699]))

print(gs[1699])
```

```python
g_mean = []
gp_mean = []
for i in g:
    g_mean.append(np.mean(i))
for i in gp:
    gp_mean.append(np.mean(i))
    

print(len(g_mean))
g_mean = np.asarray(g_mean)
fig = plt.figure(figsize=[18,9])
plt.plot(range(len(gs)), gs, label='tf_mean')
plt.plot(range(len(gs)), g_mean, label='np_mean')
plt.plot(range(len(gs)), gp_mean, label='npp_mean')
plt.legend()
```
